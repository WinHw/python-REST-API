```python
# Destructuring some values
head, *tail = [1, 2, 3, 4, 5]
print(head)
print(tail)

*head, tail = [1, 2, 3, 4, 5]
print(head)
print(tail)

head, *body, tail = [1, 2, 3, 4, 5]
print(head)
print(body)
print(tail)
```

```python
# Slicing
# syntax: [start:end]
numbers = [0, 1, 2, 3, 4, 5]

# ambil dari index 1 hingga SEBELUM index 3
print(numbers[1:3])

# ambil 3 pertama
print(numbers[:3])

# ambil dari index ke 2 hingga akhir
print(numbers[2:])

# contoh dengan negative index
print(numbers[-4:-1])
```

```python
# Conditionals

# Syntax:
# if <condition>:
#   <execution>
# elif <condition>:
#   <execution>
# ...
# else:
#   <execution>
# PERHATIKAN INDENTASI untuk block execution

# short hand if (one line)
print("OK" if 5 > 3 else "HMM")

# bila block execution tidak memiliki apapun untuk dilaksanakan
# isi dengan keyword pass
if (5 in [1, 2, 3, 4, 5]):
  pass
```

```python
# Loops pada python mirip seperti bahasa pemrograman lain

# namun while pada python juga dapat menerapkan else
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")

# for pada python juga dapat dituliskan dalam one line
codes = [f"CODE-{x}" for x in "ABCDE"]
print(codes)
```

```python
# self-defined function pada python menggunakan syntax berikut
# def function_name(arguments):
#   <execution>
#   return <value>

def print_name(firstname, lastname):
  print(firstname + " " + lastname)

def return_combined_name(firstname, lastname):
  return f"{firstname} {lastname}"

def print_names(*names):
  for name in names:
    print(name)

print_name("Hendry", "Winata")
print(return_combined_name("H","W"))
print_names("Andi", "Budi", "Cindy")

# function juga dapat menerapkan keyword pass
def noneedfunction():
  pass
```

```python
# lambda function
# small anonymous function
# take argument(s), only one expression
# syntax:
# lamda argument(s) : expression
power = lambda x, y : x ** y
print(power(2, 3))
```
